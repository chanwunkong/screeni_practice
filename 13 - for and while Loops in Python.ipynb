13 - for and while Loops in Python


1.   while and for
2.   break



```python
a = [1,2,3,4,5,]

print(a*2)
for i in a:
  print(i*2)
```

```python
count = 0
while (count<10):
  print('The count is', count)
  count = count + 1
```

```python
count = 0
while count ==0:
  print('uh oh')
```

```python
C = -40
while C<=45:
  F = (9./5)*C + 32
  print(C,F)
  # C = C +5
  C += 5
```

```python
for letter in 'I love Python':
  print('The current letter is: ', letter)
```

```python
a = ['a','b','c']
for alphabet in a:
  print('Current alphabet is', alphabet)
```

```python
for letter in 'I love python':
  if letter == 'p':
    break
  print('Current letter is ', letter)
```

```python
a = 10
while a>0:
  print('Cuttent value is', a)
  a = a-1
  if a == 5:
    break
```

```python
for num in range(10,20):  print(num)
```

```python
for i in range(10, 50, 2):  print(i)
```

```python
for num in range(10,20):
  if num%2 ==0:
    print('%d is an even number' %(num))
  else:
    print('%d is an odd number' %(num))
```
