09 - if else elif statements in Python
1.   if vs elif vs else
2.   %d vs %i vs %f
3.   逐行執行

```python
num = int(input('Please enter an integer'))

if num >= 0:
  if num == 0:
    print('zero')
  else:
    print('%i is a positive number' %(num))
elif num == 0:
  print('%d is a zero' %(num))
else:
  print('%4.2f is a positive number' %(num))
```
